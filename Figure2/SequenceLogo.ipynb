```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import logomaker
import simplejson as json
import os
plt.rcParams['svg.fonttype'] = 'none'

colors = {"DE": [0.80392, 0.12157,0.15294],
"KRH":[0.2549, 0.40392, 0.68627],
"STGCNQY":[0.39608, 0.4, 0.6],
"AVILFMWP":[0.63529,0.63137,0.45490],
         "X":[1,1,1],
         "|":[0,0,0]}

color_dict={}
for k,v in colors.items():
    for aa in k:
        color_dict[aa] = v
```

```python
nb = json.load(open("GST_morthan2_Alignment_H.json"))
hum = json.load(open("MSA_Dedup_VH_IMGT_annotated_Hum_H.json"))
mus = json.load(open("MSA_Dedup_VH_IMGT_annotated_Mus_H.json"))
```

```python
AA="GPAVLIMCFYWHKRQNEDST-"
AA2="GPAVLIMCFYWHKRQNEDSTX"
```

## Nb Database Structure

```python
nb_matrix = np.zeros((21,len(nb.keys())))
for i,item in enumerate(nb.items()):
    k,v = item
    if '-' in v.keys() and v['-']>0.9:
        continue
    for j,aa in enumerate(AA2):
        
        if aa in v.keys():
            nb_matrix[j,i] = v[aa]

```

```python
nb_df =pd.DataFrame(nb_matrix,index=[a for a in AA2]).T

fig = plt.figure(figsize=(30,5))
ax1 = fig.add_subplot(211)
logomaker.Logo(nb_df[:68],color_scheme=color_dict,vpad=.1,width=.8,ax=ax1)
ax1.set_xticks(nb_df[:68].index)
ax1.set_xticklabels(list(nb.keys())[:68],rotation=90)
ax1.set_title("Nanobody",fontsize=20)

sns.despine()
ax2 = fig.add_subplot(212)
logomaker.Logo(nb_df[68:],color_scheme=color_dict,vpad=.1,width=.8,ax=ax2)
ax2.set_xticks(nb_df[68:].index)
ax2.set_xticklabels(list(nb.keys())[68:],rotation=90)
sns.despine()

#plt.savefig("GST_Db.svg",format='svg')
```
